## Categorical Plots : 

```python
# Creating a column that splits the total rent into different categories

sns.histplot(df['Total_Rent'], bins = 80, kde = True)
plt.xlim(0,2000)

```

```python
df.loc[(df['Total_Rent'] >= 250) & (df['Total_Rent'] < 550), 'Rent category']= 'Low rent'
```

```python
df.loc[(df['Total_Rent'] >= 550) & (df['Total_Rent'] < 1500), 'Rent category'] = 'Mid rent'
```

```python
df.loc[df['Total_Rent'] > 1500, 'Rent category'] = 'High rent'
```

```python
df['Rent category'].value_counts()
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the rent categories created above

sns.set(style="ticks")
h = sns.catplot(x="Lift", y="Total_Rent", hue="Rent category", data=df)
#plt.ylim(0,2000)
```

```python
h.savefig("Lift.png")
```

It is interesting to see that few apartments that have no lift are also equally expensive as the ones with lifts. </br>
Most of the apartments with lifts fall in the 'High rent' category.

```python
# Create a categorical plot in seaborn using the rent categories created above

sns.set(style="ticks")
h = sns.catplot(x="Has_Kitchen", y="Total_Rent", hue="Rent category", data=df)
#plt.ylim(0,3000)
```

```python
h.savefig("Kitchen.png")
```

The trend is very similar to those of apartments with lifts. </br>
The number of apartments with kitchens in the 'high rent' category are more.

```python
# Create a categorical plot in seaborn using the rent categories created above

sns.set(style="ticks")
h = sns.catplot(x="Balcony", y="Total_Rent", hue="Rent category", data=df)
#plt.ylim(0,3000)
```

```python
h.savefig("Balcony.png")
```

The rent of apartments with balconies are high. 

```python
# Create a categorical plot in seaborn using the rent categories created above

sns.set(style="ticks")
h = sns.catplot(x="State", y="Total_Rent", hue="Rent category", data=df)
h.set_xticklabels(rotation=90)
plt.ylim(0,3000)
```

```python
h.savefig("State.png")
```
